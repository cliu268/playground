```python
covid_df['total_cases'] = covid_df.new_cases.cumsum()
```

```python
# Merging data from multiple sources
# Get some more data for location info
urlretrieve('https://hub.jovian.ml/wp-content/uploads/2020/09/locations.csv', 'locations.csv')
```

```python
locations_df = pd.read_csv('locations.csv')
```

```python
locations_df
```

```python
locations_df[locations_df.location == 'Italy']
```

```python
# Add a column called location to covid_df so we have the same location column for merging later
covid_df['location'] = 'Italy'
```

```python
covid_df
```

```python
merged_df = covid_df.merge(locations_df, on="location")
```

```python
merged_df
```

```python
merged_df['cases_per_million'] = merged_df.total_cases * 1e6 / merged_df.population
```

```python
merged_df['deaths_per_million'] = merged_df.total_deaths * 1e6 / merged_df.population
```

```python
merged_df['tests_per_million'] = merged_df.total_tests * 1e6 / merged_df.population
```

```python
merged_df
```

```python
result_df = merged_df[['date',
                       'new_cases', 
                       'total_cases', 
                       'new_deaths', 
                       'total_deaths', 
                       'new_tests', 
                       'total_tests', 
                       'cases_per_million', 
                       'deaths_per_million', 
                       'tests_per_million']]
```

```python
result_df
```

```python
result_df.to_csv('results.csv', index=None)
```
